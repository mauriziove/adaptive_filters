# The Z-Transform

## 1. Definition of the Z-Transform

The Z-transform is the discrete-time equivalent of the Laplace transform. 
It converts a discrete-time signal into a complex frequency-domain representation.

### 1.1 Bilateral (Two-Sided) Z-Transform

For a sequence $x[n]$:

$$X(z) = \mathcal{Z}\{x[n]\} = \sum_{n=-\infty}^{+\infty} x[n] \cdot z^{-n}$$

where $z$ is a complex variable.

### 1.2 Unilateral (One-Sided) Z-Transform

For causal sequences and systems with initial conditions:

$$X'(z) = \mathcal{Z}'\{x[n]\} = \sum_{n=0}^{+\infty} x[n] \cdot z^{-n}$$

> **Note:** The two definitions coincide when $x[n] = 0$ for all $n < 0$ (causal sequence).

### 1.3 Relationship with DTFT

When $z = e^{j\omega}$ (unit circle), the Z-transform becomes the **Discrete-Time Fourier Transform (DTFT)**:

$$X(e^{j\omega}) = \sum_{n=-\infty}^{+\infty} x[n] \cdot e^{-j\omega n}$$

This relationship exists **only if** the unit circle $|z| = 1$ is included in the ROC.

### 1.4 Linearity Property

$$\mathcal{Z}\{a \cdot x_1[n] + b \cdot x_2[n]\} = a \cdot X_1(z) + b \cdot X_2(z)$$

## 2. Region of Convergence (ROC)

The ROC is the set of values of $z$ for which the Z-transform converges (sum is finite).

### 2.1 Condition for Convergence

The Z-transform converges absolutely if:

$$\sum_{n=-\infty}^{+\infty} |x[n] \cdot z^{-n}| < +\infty$$

Writing $z = |z|e^{j\omega}$:

$$X(z) = \sum_{n=-\infty}^{+\infty} \left(x[n] \cdot |z|^{-n}\right) \cdot e^{-j\omega n}$$

Convergence requires absolute summability of the modified sequence.

### 2.2 Uniform Convergence of Fourier Transform vs Z-Transform

| Transform | Condition for Uniform Convergence |
|-----------|----------------------------------|
| **DTFT** | $\sum_{n=-\infty}^{+\infty} |x[n]| < \infty$ (absolutely summable) |
| **Z-Transform** | $\sum_{n=-\infty}^{+\infty} |x[n] \cdot z^{-n}| < \infty$ |

**Key insight:** Due to the exponential weighting $z^{-n}$, the Z-transform may converge even when the DTFT does not.

**Example:** $x[n] = u[n]$ (unit step) is NOT absolutely summable, but $x[n] \cdot z^{-n}$ IS absolutely summable for $|z| > 1$.

### 2.3 ROC Properties

1. **The ROC is always an annular region** in the z-plane: $R_{x-} < |z| < R_{x+}$
   - $R_{x-}$ can be 0 (includes origin)
   - $R_{x+}$ can be $\infty$ (extends to infinity)

2. **The Z-transform is an analytic function** inside the ROC (all derivatives exist and are continuous).

3. **For finite-duration sequences:**
   - ROC = entire z-plane (possibly excluding $z=0$ or $z=\infty$)

4. **For right-sided sequences:**
   - ROC extends outward from the outermost pole: $|z| > |z_{max}|$

5. **For left-sided sequences:**
   - ROC extends inward from the innermost pole: $|z| < |z_{min}|$

6. **For two-sided sequences:**
   - ROC is an annulus: $|z_1| < |z| < |z_2|$ (if it exists)

7. **The ROC is always a connected region** (no gaps).

8. **The DTFT exists if and only if** the ROC includes the unit circle $|z| = 1$.

### 2.4 ROC and Poles/Zeros

For rational Z-transforms $X(z) = \frac{N(z)}{D(z)}$:
- **Zeros:** Values of $z$ where $X(z) = 0$ (roots of $N(z)$)
- **Poles:** Values of $z$ where $X(z) \to \infty$ (roots of $D(z)$)

The ROC is bounded by poles and **cannot contain any poles**.



## 3. Notable Z-Transforms <a id='3-notable'></a>

### 3.1 Unit Impulse $\delta[n]$

$$\mathcal{Z}\{\delta[n]\} = 1, \quad \text{ROC: all } z$$

### 3.2 Unit Step $u[n]$

$$\mathcal{Z}\{u[n]\} = \frac{1}{1 - z^{-1}} = \frac{z}{z-1}, \quad \text{ROC: } |z| > 1$$

### 3.3 Anti-causal Unit Step $-u[-n-1]$

$$\mathcal{Z}\{-u[-n-1]\} = \frac{1}{1 - z^{-1}} = \frac{z}{z-1}, \quad \text{ROC: } |z| < 1$$

> **Important:** Same transform as $u[n]$ but with DIFFERENT ROC! This demonstrates why specifying the ROC is essential.

### 3.4 Exponential Sequence $a^n u[n]$

$$\mathcal{Z}\{a^n u[n]\} = \frac{1}{1 - a z^{-1}} = \frac{z}{z-a}, \quad \text{ROC: } |z| > |a|$$

### 3.5 Anti-causal Exponential $-a^n u[-n-1]$

$$\mathcal{Z}\{-a^n u[-n-1]\} = \frac{1}{1 - a z^{-1}} = \frac{z}{z-a}, \quad \text{ROC: } |z| < |a|$$

### 3.6 Ramp Sequence $n a^n u[n]$

$$\mathcal{Z}\{n a^n u[n]\} = \frac{a z^{-1}}{(1 - a z^{-1})^2} = \frac{a z}{(z-a)^2}, \quad \text{ROC: } |z| > |a|$$

### 3.7 Summary Table

| Sequence | Z-Transform | ROC |
|----------|-------------|-----|
| $\delta[n]$ | $1$ | All $z$ |
| $u[n]$ | $\frac{1}{1-z^{-1}}$ | $|z| > 1$ |
| $-u[-n-1]$ | $\frac{1}{1-z^{-1}}$ | $|z| < 1$ |
| $a^n u[n]$ | $\frac{1}{1-az^{-1}}$ | $|z| > |a|$ |
| $-a^n u[-n-1]$ | $\frac{1}{1-az^{-1}}$ | $|z| < |a|$ |
| $n a^n u[n]$ | $\frac{az^{-1}}{(1-az^{-1})^2}$ | $|z| > |a|$ |
| $-n a^n u[-n-1]$ | $\frac{az^{-1}}{(1-az^{-1})^2}$ | $|z| < |a|$ |

## 4. Properties of the Z-Transform <a id='4-properties'></a>

### 4.1 Linearity

$$a \cdot x_1[n] + b \cdot x_2[n] \longleftrightarrow a \cdot X_1(z) + b \cdot X_2(z)$$

$$\text{ROC} \supseteq \text{ROC}(X_1) \cap \text{ROC}(X_2)$$

### 4.2 Time Shifting

$$x[n - n_0] \longleftrightarrow z^{-n_0} \cdot X(z)$$

$$\text{ROC} = \text{ROC}(X) \text{ (except possibly } z=0 \text{ or } z=\infty)$$

For the **unilateral** transform (important for initial conditions):

$$\mathcal{Z}'\{x[n-n_0]\} = z^{-n_0} X'(z) + \sum_{k=1}^{n_0} x[-k] \cdot z^{-n_0+k}$$

### 4.3 Multiplication by Exponential Sequence

$$z_0^n \cdot x[n] \longleftrightarrow X\left(\frac{z}{z_0}\right)$$

$$\text{ROC} = |z_0| \cdot \text{ROC}(X) \quad \text{(scaled)}$$

**Special cases:**
- If $z_0 = e^{j\omega_0}$: frequency shift (modulation)
- If $z_0 = r$ real: scaling in z-plane

### 4.4 Differentiation in Z

$$n \cdot x[n] \longleftrightarrow -z \frac{dX(z)}{dz}$$

$$\text{ROC} = \text{ROC}(X) \text{ (except possibly } z=0 \text{ or } z=\infty)$$

### 4.5 Complex Conjugation

$$x^*[n] \longleftrightarrow X^*(z^*)$$

$$\text{ROC} = \text{ROC}(X)$$

### 4.6 Time Reversal

$$x[-n] \longleftrightarrow X\left(\frac{1}{z}\right)$$

$$\text{ROC} = \frac{1}{\text{ROC}(X)}$$

(If ROC was $R_1 < |z| < R_2$, new ROC is $1/R_2 < |z| < 1/R_1$)

### 4.7 Convolution of Sequences

$$x[n] * y[n] \longleftrightarrow X(z) \cdot Y(z)$$

$$\text{ROC} \supseteq \text{ROC}(X) \cap \text{ROC}(Y)$$

### 4.8 Initial Value Theorem

For a causal sequence ($x[n] = 0$ for $n < 0$):

$$x[0] = \lim_{z \to \infty} X(z)$$

### 4.9 Final Value Theorem

If $x[n]$ is causal and $X(z)$ has no poles outside the unit circle (except possibly a single pole at $z=1$):

$$\lim_{n \to \infty} x[n] = \lim_{z \to 1} (z-1) \cdot X(z)$$

## 5. Inverse Z-Transform <a id='5-inverse'></a>

Given $X(z)$ and its ROC, we want to find $x[n]$.

### 5.1 Contour Integral (Formal Definition)

Using the **Cauchy integral theorem**:

$$\frac{1}{2\pi j} \oint_C z^{-k} dz = \begin{cases} 1 & k = 1 \ 0 & k \neq 1 \end{cases}$$

The inverse Z-transform is:

$$\boxed{x[n] = \frac{1}{2\pi j} \oint_C X(z) \cdot z^{n-1} dz}$$

where $C$ is a closed contour inside the ROC, encircling the origin counter-clockwise.

### 5.2 Substitution $z = e^{j\omega}$

If the ROC includes the unit circle:

$$x[n] = \frac{1}{2\pi} \int_{-\pi}^{+\pi} X(e^{j\omega}) \cdot e^{j\omega n} d\omega$$

This is the **inverse DTFT**.

### 5.3 Method 1: Inspection (for Simple Rational Functions)

Recognize standard forms directly.

**Example:**
$$X(z) = \frac{5}{1 - az^{-1}}, \quad |z| > |a|$$

By inspection: $x[n] = 5 a^n u[n]$

### 5.4 Method 2: Power Series Expansion (Long Division)

Expand $X(z)$ as a power series in $z^{-1}$ (for right-sided) or $z$ (for left-sided):

$$X(z) = \sum_{n=-\infty}^{+\infty} x[n] z^{-n} = \cdots + x[-1]z + x[0] + x[1]z^{-1} + x[2]z^{-2} + \cdots$$

**Example:**
$$\frac{1}{1 - az^{-1}} = 1 + az^{-1} + a^2z^{-2} + a^3z^{-3} + \cdots$$

By comparison: $x[n] = a^n u[n]$

### 5.5 Method 3: Partial Fraction Expansion

For rational functions $X(z) = \frac{N(z)}{D(z)}$:

1. If degree of numerator $\geq$ degree of denominator, perform polynomial long division first
2. Factor denominator: $D(z) = (z-p_1)(z-p_2)\cdots(z-p_N)$
3. Expand into partial fractions
4. Use standard transform pairs to invert each term

**Example:**
$$X(z) = \frac{z(z-2)}{(z-3)(z+1)}$$

Poles at $z = 3$ and $z = -1$. Three possible ROCs:
1. $|z| < 1$: left-sided sequence
2. $1 < |z| < 3$: two-sided sequence
3. $|z| > 3$: right-sided sequence


## 6. Parseval's Theorem

Parseval's theorem relates the energy in the time domain to the energy in the frequency domain.

### 6.1 General Form

$$\sum_{n=-\infty}^{+\infty} x_1[n] \cdot x_2^*[n] = \frac{1}{2\pi j} \oint_C X_1(v) \cdot X_2^*\left(\frac{1}{v^*}\right) v^{-1} dv$$

where $C$ is in the intersection of ROCs.

### 6.2 Energy Conservation (Special Case)

When $x_1[n] = x_2[n] = x[n]$:

$$\sum_{n=-\infty}^{+\infty} |x[n]|^2 = \frac{1}{2\pi j} \oint_C X(v) \cdot X^*\left(\frac{1}{v^*}\right) v^{-1} dv$$

On the unit circle ($v = e^{j\omega}$):

$$\sum_{n=-\infty}^{+\infty} |x[n]|^2 = \frac{1}{2\pi} \int_{-\pi}^{\pi} |X(e^{j\omega})|^2 d\omega$$

This shows that the total energy can be computed either in time or frequency domain.

### 6.3 Complex Convolution Theorem

The Z-transform of a product of sequences is related to a complex convolution in the z-domain:

$$x_1[n] \cdot x_2[n] \longleftrightarrow \frac{1}{2\pi j} \oint_C X_1(v) \cdot X_2\left(\frac{z}{v}\right) v^{-1} dv$$

where $C$ is in the ROC intersection.

## 7. Solving Linear Difference Equations

The unilateral Z-transform is particularly useful for solving linear constant-coefficient difference equations with initial conditions.

### 7.1 Problem Setup

Consider the general difference equation:

$$\sum_{k=0}^{N} a_k \cdot y[n-k] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$

**Assumptions:**
- Causal system
- Input specified for $n \geq 0$ (causal sequence)
- Initial conditions $y[-N], y[-N+1], \ldots, y[-1]$ are known

### 7.2 Solution Using Unilateral Z-Transform

Taking the unilateral Z-transform of both sides:

$$\sum_{k=0}^{N} a_k z^{-k} Y'(z) + \sum_{k=0}^{N-1} c_k z^{-k} = \sum_{k=0}^{M} b_k z^{-k} X'(z)$$

where $c_k$ are constants derived from initial conditions.

Solving for $Y'(z)$:

$$Y'(z) = \underbrace{-\frac{\sum_{k=0}^{N-1} c_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}}}_{\text{Zero-input response}} + \underbrace{\frac{\sum_{k=0}^{M} b_k z^{-k}}{\sum_{k=0}^{N} a_k z^{-k}} X'(z)}_{\text{Zero-state response}}$$

### 7.3 Components of the Solution

1. **Zero-Input Response (ZIR):** Output due to initial conditions only (with $x[n] = 0$)
2. **Zero-State Response (ZSR):** Output due to input only (with zero initial conditions)

The total response is the sum: $y[n] = y_{zi}[n] + y_{zs}[n]$

### 7.4 Example: First-Order System

$$y[n] - a \cdot y[n-1] = x[n], \quad y[-1] = y_0$$

Taking unilateral Z-transform:

$$Y'(z) - a(z^{-1}Y'(z) + y[-1]) = X'(z)$$

$$Y'(z)(1 - az^{-1}) = X'(z) + ay_0$$

$$Y'(z) = \underbrace{\frac{ay_0}{1-az^{-1}}}_{\text{ZIR}} + \underbrace{\frac{X'(z)}{1-az^{-1}}}_{\text{ZSR}}$$
