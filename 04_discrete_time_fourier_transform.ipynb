# Discrete-Time Fourier Transform (DTFT)

## 1. Frequency Response of LTI Systems

### Fundamental Property of LTI Systems

A key property of **Linear Time-Invariant (LTI)** systems is their response to complex exponential (sinusoidal) inputs.

> **Property:** If the input to an LTI system is a complex sinusoid (pure complex exponential) of frequency $\omega$, the output is also a complex sinusoid of the **same frequency** $\omega$, but with its amplitude and phase modified by the system.

Mathematically, if the input is:

$$x[n] = e^{j\omega n}, \quad -\infty < n < +\infty$$

Then the output $y[n]$ is obtained through convolution:

$$y[n] = h[n] * x[n] = \sum_{k=-\infty}^{+\infty} h[k] \cdot x[n-k] = \sum_{k=-\infty}^{+\infty} h[k] \cdot e^{j\omega(n-k)}$$

Factoring out $e^{j\omega n}$:

$$y[n] = e^{j\omega n} \underbrace{\left\{ \sum_{k=-\infty}^{+\infty} h[k] \cdot e^{-j\omega k} \right\}}_{H(e^{j\omega})} = H(e^{j\omega}) \cdot e^{j\omega n}$$

### The Frequency Response $H(e^{j\omega})$

The quantity $H(e^{j\omega})$ is called the **frequency response** (or **transfer function**) of the system:

$$\boxed{H(e^{j\omega}) = \sum_{k=-\infty}^{+\infty} h[k] \cdot e^{-j\omega k}}$$

This is exactly the **DTFT of the impulse response** $h[n]$.

**Key characteristics of $H(e^{j\omega})$:**
- It is a **complex function** of the real variable $\omega$
- It is a **continuous function** of $\omega$
- It is **periodic** in $\omega$ with period $2\pi$

### Amplitude and Phase Response

The frequency response can be expressed in polar form:

$$H(e^{j\omega}) = |H(e^{j\omega})| \cdot e^{j\angle H(e^{j\omega})}$$

- **$|H(e^{j\omega})|$** → **Amplitude Response** (magnitude response)
- **$\angle H(e^{j\omega})$** → **Phase Response** (principal value)

## 2. The DTFT: Analysis and Synthesis Equations

### Continuous-time analogy

For an analog signal $f(t)$, the Fourier transform pair is:

$$
\begin{cases}
F(j\omega) = \displaystyle\int_{-\infty}^{+\infty} f(t)\, e^{-j\omega t}\, dt & \text{(analysis / forward)} \\[6pt]
f(t) = \dfrac{1}{2\pi}\displaystyle\int_{-\infty}^{+\infty} F(j\omega)\, e^{j\omega t}\, d\omega & \text{(synthesis / inverse)}
\end{cases}
$$

### Discrete-time case

For a discrete sequence $x[n]$, the **Discrete-Time Fourier Transform (DTFT)** pair is:

$$
\begin{cases}
X(e^{j\omega}) = \displaystyle\sum_{n=-\infty}^{+\infty} x[n]\, e^{-j\omega n} & \text{(analysis / forward)} \\[8pt]
x[n] = \dfrac{1}{2\pi}\displaystyle\int_{-\pi}^{+\pi} X(e^{j\omega})\, e^{j\omega n}\, d\omega & \text{(synthesis / inverse)}
\end{cases}
$$

**Important properties:**

- $X(e^{j\omega})$ is **continuous** and **periodic with period $2\pi$** in $\omega$. The sum runs over a discrete index
  $n$ (no aliasing issue there); the *periodicity in $\omega$* is a direct consequence of $e^{-j\omega n}$ being periodic
  in $\omega$ with period $2\pi$ for integer $n$.
- $|X(e^{j\omega})|$ — **amplitude spectrum** of $x[n]$.
- $\angle X(e^{j\omega})$ — **phase spectrum** of $x[n]$ (principal value).

Because $X(e^{j\omega})$ is $2\pi$-periodic, it is fully described by any interval of length $2\pi$, conventionally
$-\pi \le \omega \le \pi$ (or equivalently $0 \le \omega \le 2\pi$).


## 3. Existence of the Transform

The DTFT does not always converge in the ordinary sense. There are three regimes:

**1) Absolutely summable sequences**

If

$$
\sum_{n=-\infty}^{+\infty} |x[n]| < +\infty
$$

then $X(e^{j\omega})$ exists and is a **continuous** function of $\omega$. This is a **sufficient** (not necessary)
condition, and convergence of the defining sum is **uniform**.

**2) Square-summable (finite energy) sequences**

If $x[n]$ is not absolutely summable but

$$
\sum_{n=-\infty}^{+\infty} |x[n]|^2 < +\infty
$$

then $X(e^{j\omega})$ still exists, but the sum converges only in the **mean-square sense**, not uniformly. As a
consequence, the limiting function $X(e^{j\omega})$ *may* exhibit discontinuities at isolated frequencies (the classic
example being the ideal lowpass filter, §8). The convergence is **not uniform**.

**3) Neither absolutely nor square summable**

If $x[n]$ is neither absolutely nor square summable, $X(e^{j\omega})$ can still exist in special cases, typically by
allowing impulses (Dirac deltas) in the frequency domain.

**Example:** the everlasting complex exponential

$$
x[n] = e^{j\omega_0 n}, \qquad -\infty < n < +\infty
$$

has DTFT

$$
X(e^{j\omega}) = \sum_{k=-\infty}^{+\infty} 2\pi\, \delta(\omega - \omega_0 + 2\pi k)
$$

where $\delta(\cdot)$ is the **continuous-time (Dirac) unit impulse**. This is an infinite train of impulses located at
$\omega = \omega_0 + 2\pi k$ for every integer $k$ (the periodic repetition reflects the $2\pi$-periodicity of the DTFT).


## 4. LTI Systems and the DTFT

The relation $H(e^{j\omega}) = \sum_k h[k]e^{-j\omega k}$ shows that:

$$
\text{frequency response} \;\equiv\; \text{DTFT of the impulse response}
$$

Two useful facts:

- A **stable** system (i.e. $\sum_k |h[k]| < \infty$, BIBO stability) always has a frequency response (condition 1 in §3 is automatically satisfied).
- An **FIR** system always has a frequency response, provided $|h[k]| < \infty$ for all $k$ (finite sum of finite terms).

### Output of an LTI system to a generic input

Or Convolution Theorem in the Frequency Domain

For a generic input $x[n]$ with DTFT $X(e^{j\omega})$:

$$
y[n] = \sum_{k=-\infty}^{+\infty} h[k]\, x[n-k]
= \sum_{k=-\infty}^{+\infty} h[k] \cdot \frac{1}{2\pi}\int_{-\pi}^{\pi} X(e^{j\omega})\, e^{j\omega(n-k)}\, d\omega
$$

Swapping the order of summation and integration:

$$
y[n] = \frac{1}{2\pi}\int_{-\pi}^{\pi}
\left[ \sum_{k=-\infty}^{+\infty} h[k]\, e^{-j\omega k} \right] X(e^{j\omega})\, e^{j\omega n}\, d\omega
= \frac{1}{2\pi}\int_{-\pi}^{+\pi} H(e^{j\omega})\, X(e^{j\omega})\, e^{j\omega n}\, d\omega
$$

But by the inverse DTFT, $y[n] = \dfrac{1}{2\pi}\displaystyle\int_{-\pi}^{\pi} Y(e^{j\omega})\, e^{j\omega n}\, d\omega$,
so by uniqueness of the DTFT:

$$
\boxed{Y(e^{j\omega}) = H(e^{j\omega}) \cdot X(e^{j\omega})}
$$

**Convolution in time corresponds to multiplication in frequency** — exactly as in continuous-time Fourier analysis.


## 5. Properties of the DTFT

| Property | Time domain | DTFT domain |
|---|---|---|
| Linearity | $a\,x_1[n] + b\,x_2[n]$ | $a\,X_1(e^{j\omega}) + b\,X_2(e^{j\omega})$ |
| Time shift | $x[n-n_0]$ | $e^{-j\omega n_0}\, X(e^{j\omega})$ |
| Frequency shift (modulation by complex exponential) | $e^{j\omega_0 n}\, x[n]$ | $X\!\left(e^{j(\omega-\omega_0)}\right)$ |
| Time reversal | $x[-n]$ | $X(e^{-j\omega})$  — equals $X^*(e^{j\omega})$ if $x[n]$ is **real** |
| Convolution | $x[n] * y[n]$ | $X(e^{j\omega})\, Y(e^{j\omega})$ |
| Modulation (multiplication in time) | $x[n]\, y[n]$ | $\dfrac{1}{2\pi}\displaystyle\int_{-\pi}^{+\pi} X(e^{j\theta})\, Y\!\left(e^{j(\omega-\theta)}\right) d\theta$  (periodic convolution) |

### Parseval's theorem

$$
\sum_{n=-\infty}^{+\infty} |x[n]|^2 = \frac{1}{2\pi}\int_{-\pi}^{+\pi} \left|X(e^{j\omega})\right|^2 d\omega
\qquad \text{(energy of a sequence)}
$$

$\left|X(e^{j\omega})\right|^2$ is called the **energy density spectrum**.

More generally, for the cross-energy of two sequences:

$$
\sum_{n=-\infty}^{+\infty} x[n]\, y^*[n] = \frac{1}{2\pi}\int_{-\pi}^{+\pi} X(e^{j\omega})\, Y^*(e^{j\omega})\, d\omega
$$


## 6. Symmetry Properties of the DTFT

The symmetry properties are useful in practice: they let you infer properties of the spectrum from properties of the
signal (and vice versa) without recomputing the whole transform.

### Conjugate-symmetric / conjugate-antisymmetric decomposition

Any sequence $x[n]$ can be decomposed as $x[n] = x_e[n] + x_o[n]$, where:

- $x_e[n]$ is **conjugate-symmetric**: $x_e[n] = x_e^*[-n]$, obtained as
$$x_e[n] = \frac{1}{2}\Big(x[n] + x^*[-n]\Big)$$
- $x_o[n]$ is **conjugate-antisymmetric**: $x_o[n] = -x_o^*[-n]$, obtained as
$$x_o[n] = \frac{1}{2}\Big(x[n] - x^*[-n]\Big)$$

**Special case — real sequences:**

- If $x[n]$ is real and **even** ($x[n] = x[-n]$), this is the special case of $x_e[n]$.
- If $x[n]$ is real and **odd** ($x[n] = -x[-n]$), this is the special case of $x_o[n]$.

### General duality relations

These two relations are the building blocks for everything else in this section:

$$
x^*[n] \;\longleftrightarrow\; X^*(e^{-j\omega}) \qquad\qquad x^*[-n] \;\longleftrightarrow\; X^*(e^{j\omega})
$$

From these it follows that:

$$
\mathrm{Re}\{x[n]\} \;\longleftrightarrow\; X_e(e^{j\omega}) \qquad\qquad
x_e[n] \;\longleftrightarrow\; \mathrm{Re}\{X(e^{j\omega})\}
$$

$$
j\,\mathrm{Im}\{x[n]\} \;\longleftrightarrow\; X_o(e^{j\omega}) \qquad\qquad
x_o[n] \;\longleftrightarrow\; j\,\mathrm{Im}\{X(e^{j\omega})\}
$$

*(In the original handwritten notes, page 11 lists $x^*[-n] \leftrightarrow X^*(e^{j\omega})$ next to the symmetric-decomposition
results — that pairing is correct as written above; the layout there can read ambiguously, so it's spelled out explicitly here.)*

### For real sequences — the case that matters most in practice

If $x[n]$ is **real**, then $X(e^{j\omega})$ satisfies **conjugate symmetry**:

$$
\boxed{X(e^{j\omega}) = X^*(e^{-j\omega})}
$$

which implies:

| Quantity | Symmetry |
|---|---|
| $\mathrm{Re}\{X(e^{j\omega})\}$ | **Even** function of $\omega$ |
| $\mathrm{Im}\{X(e^{j\omega})\}$ | **Odd** function of $\omega$ |
| $\lvert X(e^{j\omega}) \rvert$ | **Even** function of $\omega$ |
| $\angle X(e^{j\omega})$ | **Odd** function of $\omega$ |

> **Practical consequence:** for a real sequence it is enough to know the transform on **half the interval**,
> $0 \le \omega \le \pi$ — the rest follows by symmetry. This is why magnitude plots are conventionally drawn only for
> $\omega \in [0, \pi]$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Demonstration: conjugate symmetry of the DTFT for a real sequence
# x[n] is a short real, non-symmetric sequence
n = np.arange(-3, 4)
x = np.array([0.0, 0.5, 1.0, 2.0, -0.5, 0.3, 0.0])  # real, not even/odd

omega = np.linspace(-np.pi, np.pi, 1000)
X = np.array([np.sum(x * np.exp(-1j * w * n)) for w in omega])

fig, axs = plt.subplots(2, 2, figsize=(11, 6))

axs[0, 0].plot(omega, X.real)
axs[0, 0].set_title(r"$\mathrm{Re}\{X(e^{j\omega})\}$ — even")
axs[0, 0].axvline(0, color='gray', lw=0.5)
axs[0, 0].set_xlabel(r"$\omega$")

axs[0, 1].plot(omega, X.imag, color='orange')
axs[0, 1].set_title(r"$\mathrm{Im}\{X(e^{j\omega})\}$ — odd")
axs[0, 1].axvline(0, color='gray', lw=0.5)
axs[0, 1].set_xlabel(r"$\omega$")

axs[1, 0].plot(omega, np.abs(X), color='green')
axs[1, 0].set_title(r"$|X(e^{j\omega})|$ — even")
axs[1, 0].axvline(0, color='gray', lw=0.5)
axs[1, 0].set_xlabel(r"$\omega$")

axs[1, 1].plot(omega, np.angle(X), color='red')
axs[1, 1].set_title(r"$\angle X(e^{j\omega})$ — odd")
axs[1, 1].axvline(0, color='gray', lw=0.5)
axs[1, 1].set_xlabel(r"$\omega$")

plt.suptitle("Symmetry of the DTFT for a real sequence $x[n]$")
plt.tight_layout()
plt.show()
```

## 7. A Few Common DTFT Pairs

| $x[n]$ | $X(e^{j\omega})$ |
|---|---|
| $\delta[n]$ | $1$ |
| $\delta[n-n_0]$ | $e^{-j\omega n_0}$ |
| $1$  (for all $n$) | $\displaystyle\sum_{k=-\infty}^{+\infty} 2\pi\,\delta(\omega - 2\pi k)$ |
| $e^{j\omega_0 n}$ | $\displaystyle\sum_{k=-\infty}^{+\infty} 2\pi\,\delta(\omega - \omega_0 + 2\pi k)$ |
| $\dfrac{\sin(\omega_c n)}{\pi n}$ | $\begin{cases}1 & \lvert\omega\rvert < \omega_c \\ 0 & \omega_c < \lvert\omega\rvert \le \pi\end{cases}$ |
| $x[n] = \begin{cases}1 & 0\le n\le M\\ 0 & \text{otherwise}\end{cases}$ | $\dfrac{\sin\!\left[\omega(M+1)/2\right]}{\sin(\omega/2)}\, e^{-j\omega M/2}$ |
| $a^n u[n]$, $\lvert a \rvert < 1$ | $\dfrac{1}{1 - a\,e^{-j\omega}}$ |

> **Duality Note:** There is a duality between time and frequency domains. A narrow signal in time produces a wide spectrum, and vice versa. This is illustrated below.


```python
# Visualize common DTFT pairs

fig, axes = plt.subplots(2, 3, figsize=(14, 8))

# --- Pair 1: delta[n] <-> 1 ---
n = np.arange(-10, 11)
x1 = np.zeros_like(n, dtype=float)
x1[n == 0] = 1
axes[0, 0].stem(n, x1, basefmt=' ')
axes[0, 0].set_title('x[n] = delta[n]')
axes[0, 0].set_xlabel('n')
axes[0, 0].grid(True, alpha=0.3)

omega = np.linspace(-np.pi, np.pi, 200)
X1 = np.ones_like(omega)
axes[1, 0].plot(omega, X1, 'b-')
axes[1, 0].set_title('X(e^{jomega}) = 1')
axes[1, 0].set_xlabel('omega')
axes[1, 0].grid(True, alpha=0.3)

# --- Pair 2: Rectangular window <-> Dirichlet kernel ---
M = 5
x2 = np.where((n >= 0) & (n <= M), 1.0, 0.0)
axes[0, 1].stem(n, x2, basefmt=' ')
axes[0, 1].set_title(f'x[n] = 1 for 0 leq n leq {M}')
axes[0, 1].set_xlabel('n')
axes[0, 1].grid(True, alpha=0.3)

X2 = np.sin(omega * (M+1) / 2) / (np.sin(omega / 2) + 1e-10)
axes[1, 1].plot(omega, np.abs(X2), 'b-')
axes[1, 1].set_title('|X(e^{jomega})| = Dirichlet Kernel')
axes[1, 1].set_xlabel('omega')
axes[1, 1].grid(True, alpha=0.3)

# --- Pair 3: a^n u[n] <-> 1/(1 - a*e^{-jw}) ---
a = 0.7
n_pos = np.arange(0, 21)
x3 = a ** n_pos
axes[0, 2].stem(n_pos, x3, basefmt=' ')
axes[0, 2].set_title(f'x[n] = {a}^n u[n]')
axes[0, 2].set_xlabel('n')
axes[0, 2].grid(True, alpha=0.3)

X3 = 1 / (1 - a * np.exp(-1j * omega))
axes[1, 2].plot(omega, np.abs(X3), 'b-')
axes[1, 2].set_title('|X(e^{jomega})| = |1/(1 - 0.7e^{-jomega})|')
axes[1, 2].set_xlabel('omega')
axes[1, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print('Common DTFT pairs visualized!')
```

## 8. Ideal Filters

A **filter** is a circuit/system characterized by being strongly selective in frequency.

### Ideal filters

Ideal filters are discrete systems that allow shaping the spectrum of the input signal with a "brick-wall" response.

**Lowpass (LP):**
$$
H_{lp}(e^{j\omega}) = \begin{cases}1 & \lvert\omega\rvert < \omega_c \\ 0 & \omega_c < \lvert\omega\rvert \le \pi\end{cases}
$$

- Passes frequencies below cutoff $\omega_c$
- Rejects frequencies above cutoff

**Highpass (HP):**
$$
H_{hp}(e^{j\omega}) = \begin{cases}1 & \omega_c < \lvert\omega\rvert \le \pi \\ 0 & \lvert\omega\rvert < \omega_c\end{cases}
$$

- Passes frequencies above cutoff $\omega_c$
- Rejects frequencies below cutoff


**Bandpass (BP):**
$$
H_{bp}(e^{j\omega}) = \begin{cases}1 & \omega_a < \lvert\omega\rvert < \omega_b \\ 0 & \text{otherwise in } (-\pi,\pi]\end{cases}
$$

- Passes frequencies between $\omega_a$ and $\omega_b$
- Rejects all other frequencies

**Bandstop / notch (BS):**
$$
H_{sb}(e^{j\omega}) = \begin{cases}0 & \omega_a < \lvert\omega\rvert < \omega_b \\ 1 & \text{otherwise in } (-\pi,\pi]\end{cases}
$$

All four responses are $2\pi$-periodic, as required for any DTFT.


```python
# Visualize ideal filters
omega = np.linspace(-np.pi, np.pi, 1000)
wc = np.pi / 3  # cutoff frequency
wa = np.pi / 6  # band-pass lower edge
wb = np.pi / 2  # band-pass upper edge

# Low-pass
H_lp = np.where(np.abs(omega) < wc, 1.0, 0.0)
# High-pass
H_hp = np.where(np.abs(omega) > wc, 1.0, 0.0)
# Band-pass
H_bp = np.where((np.abs(omega) > wa) & (np.abs(omega) < wb), 1.0, 0.0)
# Band-stop
H_sb = np.where((np.abs(omega) > wa) & (np.abs(omega) < wb), 0.0, 1.0)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(omega, H_lp, 'b-', linewidth=2)
axes[0, 0].set_title('Ideal Low-Pass Filter')
axes[0, 0].set_xlabel('omega')
axes[0, 0].set_ylabel('H_{lp}(e^{jomega})')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].set_ylim(-0.1, 1.2)

axes[0, 1].plot(omega, H_hp, 'r-', linewidth=2)
axes[0, 1].set_title('Ideal High-Pass Filter')
axes[0, 1].set_xlabel('omega')
axes[0, 1].set_ylabel('H_{hp}(e^{jomega})')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_ylim(-0.1, 1.2)

axes[1, 0].plot(omega, H_bp, 'g-', linewidth=2)
axes[1, 0].set_title('Ideal Band-Pass Filter')
axes[1, 0].set_xlabel('omega')
axes[1, 0].set_ylabel('H_{bp}(e^{jomega})')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_ylim(-0.1, 1.2)

axes[1, 1].plot(omega, H_sb, 'm-', linewidth=2)
axes[1, 1].set_title('Ideal Band-Stop Filter')
axes[1, 1].set_xlabel('omega')
axes[1, 1].set_ylabel('H_{sb}(e^{jomega})')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_ylim(-0.1, 1.2)

plt.tight_layout()
plt.show()

print('Ideal filter frequency responses plotted!')
```

## 9. Ideal Lowpass Filter and its Impulse Response

Taking the inverse DTFT of the ideal lowpass response:

$$
h_{lp}[n] = \frac{1}{2\pi}\int_{-\omega_c}^{+\omega_c} e^{j\omega n}\, d\omega = \frac{\sin(\omega_c n)}{\pi n},
\qquad -\infty < n < +\infty
$$

This is the **sinc** function — infinite in length and **non-causal** (it's nonzero for $n<0$). It also decays slowly
($\propto 1/n$), so it is **not absolutely summable**, only square-summable. This matches the discussion in §3: $h_{lp}[n]$
falls in convergence regime (2), and indeed the partial-sum approximations converge in the **mean-square** sense, not
uniformly:

$$
\lim_{M\to\infty}\int_{-\pi}^{+\pi} \left| H_{lp}(e^{j\omega}) - \sum_{n=-M}^{M} \frac{\sin(\omega_c n)}{\pi n}\, e^{-j\omega n}\right|^2 d\omega = 0
$$

**Conclusion: ideal filters do not exist** as realizable (causal, finite) systems — $h_{lp}[n]$ has infinite duration and
is non-causal. In practice, one truncates it to a finite-length FIR filter and then shifts it in time to make it
causal (a delay of $M$ samples, where $2M{+}1$ is the truncated length).


```python
import numpy as np
import matplotlib.pyplot as plt

wc = np.pi / 3   # cutoff frequency
M = 25            # truncation length (keep |n| <= M)

n_full = np.arange(-60, 61)
h_full = np.sinc(wc * n_full / np.pi) * wc / np.pi  # sin(wc n)/(pi n), with sinc(0)=1 handled by np.sinc

n_trunc = np.arange(-M, M + 1)
h_trunc = np.sinc(wc * n_trunc / np.pi) * wc / np.pi

fig, axs = plt.subplots(1, 2, figsize=(11, 4))
axs[0].stem(n_full, h_full, basefmt=" ")
axs[0].set_title(r"Ideal LPF impulse response $h_{lp}[n]=\sin(\omega_c n)/(\pi n)$")
axs[0].set_xlabel("n")

axs[1].stem(n_trunc, h_trunc, basefmt=" ")
axs[1].set_title(f"Truncated to $|n|\\leq{M}$ (still non-causal)")
axs[1].set_xlabel("n")

plt.tight_layout()
plt.show()
```

## 10. The Gibbs Phenomenon

If we truncate the ideal lowpass impulse response to a finite length and compute the resulting (now-realizable)
frequency response, we don't get the ideal brick wall back — we get **ripple** ("ringing") near the discontinuity at
$\omega_c$.

Key empirical facts about the Gibbs phenomenon:

- The **longer** the truncated sequence, the better the approximation gets *in terms of total error energy* (consistent
  with mean-square convergence).
- However, the **ripple does not shrink in amplitude** as the truncation length increases — it only gets
  **compressed closer to the discontinuity**. The peak overshoot stays at roughly the same height no matter how many
  terms you keep.
- This is why **ideal filters cannot be implemented exactly**: any finite-length (FIR) approximation will show ripple
  in its frequency response.

Let's see this numerically by truncating the ideal LPF impulse response at increasing lengths and comparing the
resulting frequency responses.


```python
import numpy as np
import matplotlib.pyplot as plt

wc = np.pi / 3
omega = np.linspace(-np.pi, np.pi, 2000)

def H_truncated(M, omega):
    n = np.arange(-M, M + 1)
    h = np.sinc(wc * n / np.pi) * wc / np.pi
    H = np.array([np.sum(h * np.exp(-1j * w * n)) for w in omega])
    return H

fig, axs = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
for ax, M in zip(axs, [5, 15, 50]):
    H = H_truncated(M, omega)
    ax.plot(omega, H.real)
    ax.axvline(wc, color='red', ls='--', lw=1, label=r'$\omega_c$')
    ax.axvline(-wc, color='red', ls='--', lw=1)
    ax.set_title(f"Truncated at M = {M}")
    ax.set_xlabel(r"$\omega$")
    ax.legend()

axs[0].set_ylabel(r"$H(e^{j\omega})$ (real part)")
plt.suptitle("Gibbs phenomenon: ripple compresses toward the discontinuity but does not shrink in height")
plt.tight_layout()
plt.show()
```

## Summary

- The **frequency response** $H(e^{j\omega})$ of an LTI system is the DTFT of its impulse response $h[n]$; eigenfunctions
  of LTI systems are complex exponentials.
- The **DTFT** maps a discrete-time sequence to a continuous, $2\pi$-periodic function of $\omega$; the inverse DTFT
  recovers the sequence by integrating over one period.
- Existence depends on summability: absolute summability gives uniform convergence (continuous transform);
  square summability gives mean-square convergence (transform can have discontinuities); some non-summable signals
  still have a transform if impulses are allowed.
- Convolution in time $\leftrightarrow$ multiplication in frequency: $Y(e^{j\omega}) = H(e^{j\omega})X(e^{j\omega})$.
- Real-valued sequences have conjugate-symmetric transforms, so only $\omega \in [0,\pi]$ needs to be computed/plotted.
- **Ideal filters** (LP/HP/BP/BS) are useful conceptual tools but are not realizable: their impulse responses are
  infinite and non-causal. Truncating them to a finite (FIR) length introduces the **Gibbs phenomenon** — ripple that
  does not vanish with longer truncation, only narrows around the discontinuity.
