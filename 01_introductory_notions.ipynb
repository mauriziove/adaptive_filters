# Introductory Notions — Signal Processing

This notebook covers the basics of signals and systems for signal processing: definitions, classification (analog, discrete, digital), comparison of analog vs. digital systems, and a short historical overview leading to the current state of the field.

## 1. Overview

- **Signals** can be classified as: analog, discrete, or digital.
- **Systems for signal processing** can likewise be: analog, discrete, or digital (also called *Signal Processing Systems* — circuits or algorithms).

## 2. What is a Signal?

> **Signal** — any physical quantity to which a piece of information is associated.

- **Information ⇔ Signals**
- **Manipulation of information ⇔ Systems for signal processing** (Signal Processing Systems: circuits, algorithms)

In general, a signal is a function of several independent variables. Usually we are interested in how it evolves over time:

$$\text{Signal} \;\Longleftrightarrow\; \text{Function of time}$$

**Examples**
- Voice signal = continuous, bounded function of time.
- Image = continuous, bounded function of space (2 variables: $x$, $y$).

### Signal Classification

```
                     Signals
          ┌────────────┼────────────┐
      analog        discrete       digital
```

- **Analog**: continuous functions.
- **Discrete**: discrete functions.
- **Digital**: discrete functions with limited precision.

## 3. Analog Signals

**Bounded, real, and everywhere continuous functions.**

Analog signals reproduce — i.e., are in close *analogy* with — the behavior of physical quantities of the real world.

**Example:** voltages and currents in an electrical/electronic circuit.

**Example:** functions of time, where
- $t$ = real (continuous time)
- $x(t)$ = real

Notes:
- By extension, complex analog signals are also considered.
- The constraint of range is often removed.

````python
import numpy as np
import matplotlib.pyplot as plt

# Example of an analog signal x(t): continuous function of time
t = np.linspace(0, 10, 1000)
x_t = np.sin(t) + 0.3*np.sin(3*t)

plt.figure(figsize=(7,3))
plt.plot(t, x_t, color='crimson')
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Example of an analog signal x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(alpha=0.3)
plt.show()
````

## 4. Discrete Signals

**Sequences of numbers.**

Signals defined as functions of independent variables that can only take discrete values.

**Example:** function of a single variable $n$ (time), where
- $n$ = discrete
- $x(n)$ = real

Discrete signals are often obtained by **sampling** the corresponding analog signals.

````python
# Example of a discrete signal x(n): sequence of real numbers
# Sampling an analog signal to obtain a discrete one
t_fine = np.linspace(0, 10, 1000)
x_analog = np.sin(t_fine) + 0.3*np.sin(3*t_fine)
n_sample = np.arange(0, 10, 0.5)
x_sampled = np.sin(n_sample) + 0.3*np.sin(3*n_sample)

plt.figure(figsize=(7,3))
plt.plot(t_fine, x_analog, color='lightgray', label='analog signal')
plt.stem(n_sample, x_sampled, linefmt='crimson', markerfmt='ro', basefmt='black', label='sampled (discrete) signal')
plt.title('Example of a discrete signal x(n)')
plt.xlabel('t / n')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

## 5. Digital Signals

**Sequences of digital numbers.**

Signals defined as discrete functions of discrete independent variables — i.e., **discretization of both the domain and the codomain**.

**Example:** function of a single variable $n$ (time), where
- $n$ = discrete
- $x(n)$ = discrete

**Digital numbers ⇔ Precision, Magnitude**

Various representation types:
- **Integer** (two's complement)
- **Real** (fixed-point, floating-point)

````python
# Example of a digital signal x(n): discrete domain AND quantized (discrete) amplitude
n = np.arange(0, 15)
x_continuous_amp = np.array([0.2,0.5,1.0,1.8,2.5,2.0,1.6,1.2,0.9,0.6,0.4,0.3,0.2,0.1,0.05])

# Quantize the amplitude to a limited set of levels (limited precision)
levels = np.linspace(0, 2.5, 6)  # 6 quantization levels
x_digital = levels[np.digitize(x_continuous_amp, levels) - 1]

plt.figure(figsize=(7,3))
plt.stem(n, x_digital, linefmt='crimson', markerfmt='ro', basefmt='black')
for lvl in levels:
    plt.axhline(lvl, color='gray', linewidth=0.5, linestyle='--')
plt.title('Example of a digital signal x(n) — quantized amplitude')
plt.xlabel('n')
plt.ylabel('x(n) (quantized)')
plt.grid(alpha=0.3)
plt.show()
````

## 6. Systems for Signal Processing

```
                Systems for signal processing
        ┌───────────────┬───────────────┐
   analog systems   discrete systems  digital systems
```

**Analog systems**

$$x(t) \;\longrightarrow\; \boxed{\text{analog circuit}} \;\longrightarrow\; y(t)$$
input: analog signal → output: analog signal

**Discrete systems**

$$x(n) \;\longrightarrow\; \boxed{\text{numerical circuit}} \;\longrightarrow\; y(n)$$
input: discrete signal → output: discrete signal

**Digital systems**

$$x(n) \;\longrightarrow\; \boxed{\text{digital circuit}} \;\longrightarrow\; y(n)$$
input: digital signal → output: digital signal

## 7. Analog Systems

Usually implemented with electrical circuits made of lumped or distributed constants.

**Advantages**
- High speed
- Potentially low cost
- Ability to handle large amounts of power

**Disadvantages**
- Sensitivity to noise
- Lack of reproducibility
- Poor flexibility

## 8. Digital Systems

A good approximation of discrete systems; often implemented with programmable digital circuits.

**Advantages**
- Reproducibility
- Insensitivity to noise
- High flexibility
- Low cost relative to complexity
- Possibility of integration (VLSI)
- Ease of interaction with humans

**Disadvantages**
- Low speed
- Precision issues (internal noise)
- Currently limited ability to handle power

## 9. Historical Evolution

### 1930s – 1950s
- Development of analog systems (electrical, mechanical, pneumatic, etc.). In particular, synthesis techniques were developed for linear, stationary, lumped-constant electrical circuits (synthesis of one-port and two-port networks using R, L, C, M).

### 1950s – 1960s
- First appearance of digital computers applied to the simulation of analog systems.
- Theoretical development of discrete systems starting from analog ones.

### 1965 – Present
- Discovery of the **FFT** (Fast Fourier Transform), which made it possible to apply digital systems even to real-world signals (drastic reduction in computation time).
- Development of digital and discrete systems as a field in its own right, not necessarily tied to analog systems.
- Development of techniques and tools for the interpretation of signals (**pattern recognition**).

## 10. Current Situation

**Analog systems**
- Applications at high frequency (VHF, UHF, microwave)
- Power applications (network filters, crossover filters, etc.)

**Discrete / digital systems**
- Non-real-time applications (practically in every field)
- Real-time applications (in every field permitted by speed and the development of dedicated hardware)

**Application domains:** voice signals, images, biological signals, seismic data, radar, astronomy, information transmission, music signals.

> Research and applications in **Digital Signal Processing (DSP)** have been growing rapidly for many years.

## Summary

| | Analog | Discrete | Digital |
|---|---|---|---|
| Domain | continuous | discrete | discrete |
| Amplitude (codomain) | continuous | continuous | discrete (quantized) |
| Typical implementation | electrical circuits | numerical circuits | digital/programmable circuits |
| Main strength | speed, power handling | — | flexibility, reproducibility |
| Main weakness | noise sensitivity, poor flexibility | — | speed, internal noise/precision |
